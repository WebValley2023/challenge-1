```python
import pandas as pd
```

```python
aq=pd.read_csv('Riva_final.csv')
weather=pd.read_csv('WaterHourlyAvg.csv')
```

```python
aq.rename(columns={'date': 'date'}, inplace=True)
weather.rename(columns={'Date Time': 'date'}, inplace=True)
data=pd.merge(aq, weather, on='date')
```

```python
data
```

```python
data.rename(columns={'Biossido di Azoto': 'NO2'}, inplace=True)
data.rename(columns={'Ozone': 'O3'}, inplace=True)
```

```python
data
```

```python
data.drop_duplicates(subset='date', inplace=True)
data.to_csv('Riva_final_withWater.csv')
```
