```python
#Single

Appa=pd.read_csv(
    "DataSet_Riva_garda.csv",
    na_values="n.d.",
)

Appa["Data"]=pd.to_datetime(Appa["Data"])



Appa.sort_values(by="Data", inplace=True)

Appa=Appa.set_index("Data",drop=True)
#Appa = Appa.resample("M").mean()

plt.figure(figsize=(20, 10))  
plt.plot(Appa.index, Appa["Ozone"]*5)  
plt.xlabel("Time")
plt.ylabel("Ozone")
start_data = "20/06/2019"
end_data = "22/06/2019"
plt.xlim(pd.to_datetime(start_data),pd.to_datetime(end_data))

plt.plot(Appa.index, Appa["Nitrogen dioxide"])  
plt.xlim(pd.to_datetime(start_data),pd.to_datetime(end_data))

plt.plot(Appa.index, Appa["PM10"])  
plt.xlim(pd.to_datetime(start_data),pd.to_datetime(end_data))

plt.plot(Appa.index, Appa["Air_humidity"])  
plt.xlim(pd.to_datetime(start_data),pd.to_datetime(end_data))

plt.plot(Appa.index, Appa["Sol_rad"])  
plt.xlim(pd.to_datetime(start_data),pd.to_datetime(end_data))

plt.plot(Appa.index, Appa["Temperature"])  
plt.xlim(pd.to_datetime(start_data),pd.to_datetime(end_data))

plt.plot(Appa.index, Appa["Wind_speed"])  
plt.xlim(pd.to_datetime(start_data),pd.to_datetime(end_data))

plt.legend(["Ozone","Nitrogen dioxide","PM10","Air_humidity","Sol_rad","Temperature","Wind_speed"])
plt.show()
```
