```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
lista_file = [
    "APPA_2_Parco S. Chiara.csv",
    "APPA_4_Via Bolzano.csv",
    "APPA_6_Rovereto.csv",
    "APPA_8_Borgo Valsugana.csv",
    "APPA_9_Riva del Garda.csv",
    "APPA_15_Monte Gaza.csv",
    "APPA_22_Piana Rotaliana.csv",
    "APPA_23_A22 (Avio).csv",
]
```

```python
for file in lista_file:
    Appa=pd.read_csv(
        "..\\DataSet_APPA_TRENTO\\DataSet_APPA_(2013-2023)\\"+file,
        na_values="n.d."
    )
    Appa["Data"]=pd.to_datetime(Appa["Data"])
    

    
    matrix=Appa.set_index("Data").resample("M").mean().reset_index().corr()
    matrix=matrix.drop(["Data"],axis=0)
    matrix=matrix.drop(["Data"],axis=1)

    plt.figure()  # Crea una nuova figura
    sns.heatmap(
        data=matrix,
        annot=True
    )
    matrix.style.background_gradient(cmap='coolwarm')
    plt.title("\n\n"+file+"\n\n")
    plt.show()  # Mostra la figura
    
```
